<h5> 7. Se crea una columna de decade en los df df_poblacion y df_PBI para futuros análisis</h5>

<h5> 5. Se eliminan columnas que no son necesarias </h5>

```python
df_PBI = df_PBI.drop(columns=['Indicator Name', 'Indicator Code', 'Unnamed: 65', '2020'])
```

<h5> 6. Se transforman las columnas de lo años (1960 a 2020) en una columna Years que tiene los datos de los años</h5>

```python
df_poblacion = df_poblacion.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Population')
df_poblacion['Year'] = df_poblacion['Year'].astype(int)

df_poblacion
```

```python
df_PBI = df_PBI.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='PBI')
df_PBI['Year'] = df_PBI['Year'].astype(int)

df_PBI
```

<h5> 7. Se crea una columna de decade en los df df_poblacion, df_PBI y df_JJOO para futuros análisis</h5>

```python
df_poblacion['Decade'] = (df_poblacion['Year'] // 10) * 10 
df_PBI['Decade'] = (df_PBI['Year'] // 10) * 10 
df_JJOO['Decade'] = (df_JJOO['Year'] // 10) * 10 
```

<h5> 8. Se reinicia todos los index</h5>

```python
df_poblacion = df_poblacion.reset_index(drop=True)
df_PBI = df_PBI.reset_index(drop=True)
df_JJOO = df_JJOO.reset_index(drop=True)
df_sportsman = df_sportsman.reset_index(drop=True)
```
